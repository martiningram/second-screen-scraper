```python
from bs4 import BeautifulSoup
```

```python
def get_tournament_result_links(year):
    
    page = requests.get(f'https://www.atptour.com/en/scores/results-archive?year={year}')
    
    soup = BeautifulSoup(page.content, 'html.parser')
    
    all_results = soup.find_all('tr', attrs={'class': 'tourney-result'})
    
    # This is a bit tenuous as it relies on the class of the a element...
    all_result_buttons = [x.find('a', attrs={'class': 'button-border'}) for x in all_results]
    all_result_links = [x.get('href') for x in all_result_buttons if x is not None]
    
    to_add = 'https://www.atptour.com'
    
    with_prefix = [to_add + x for x in all_result_links]
    
    return with_prefix
```
